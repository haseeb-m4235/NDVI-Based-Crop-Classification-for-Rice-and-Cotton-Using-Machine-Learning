```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Function to plot confusion matrices
def plot_confusion_matrix(conf_matrix, year):
    plt.figure(figsize=(6, 5))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=["Rice", "Cotton"], yticklabels=["Rice", "Cotton"])
    plt.title(f"Confusion Matrix for {year}")
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.show()

# Function to plot feature importances
def plot_feature_importances(feature_importances, year):
    features = [f"NDVI{i+1}" for i in range(len(feature_importances))]
    plt.figure(figsize=(10, 5))
    plt.bar(features, feature_importances)
    plt.title(f"Feature Importances for {year}")
    plt.xlabel("Features")
    plt.ylabel("Importance")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Function to plot mean scores of all years with numerical values on top
def plot_mean_scores(results):
    metrics = ["accuracy", "precision", "recall", "f1", "weighted_f1"]
    mean_scores = {metric: np.mean([result[metric] for result in results]) for metric in metrics}
    
    plt.figure(figsize=(8, 5))
    sns.barplot(x=list(mean_scores.keys()), y=list(mean_scores.values()), palette="viridis")
    plt.title("Mean Scores Across All Years")
    plt.xlabel("Metrics")
    plt.ylabel("Mean Score")
    plt.ylim(0, 1)

    # Add numerical values on top of the bars
    for i, (metric, value) in enumerate(mean_scores.items()):
        plt.text(i, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.show()
```

# Bagging Bagging (BootstrapAggregation) 

```python
# Load the JSON data
with open("C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\Bagging_best_hyperparameters_results.json", "r") as file:
    results = json.load(file)

# Iterate over results and generate visualizations
for result in results:
    year = result["year"]
    print(f"Results for {year}:")
    print(f"Classification Report:\n{result['classification_report']}\n")
    
    # Confusion matrix
    plot_confusion_matrix(np.array(result["confusion_matrix"]), year)
    
    # Feature importances
    plot_feature_importances(result["feature_importances"], year)

# Plot mean scores across all years
plot_mean_scores(results)
```

# Models comparison on mean scores of all year

```python
import json
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Function to calculate mean scores for a single model
def calculate_mean_scores(file_path):
    with open(file_path, "r") as file:
        results = json.load(file)
    metrics = ["accuracy", "precision", "recall", "f1", "weighted_f1"]
    mean_scores = {metric: np.mean([result[metric] for result in results]) for metric in metrics}
    return mean_scores

# Function to visualize comparison of models
def compare_models(file_paths, model_names):
    all_mean_scores = {model: calculate_mean_scores(file) for model, file in zip(model_names, file_paths)}
    
    metrics = list(next(iter(all_mean_scores.values())).keys())  # Get metrics from the first model
    mean_scores_matrix = np.array([list(mean_scores.values()) for mean_scores in all_mean_scores.values()])
    
    # Plotting the comparison
    plt.figure(figsize=(10, 6))
    x = np.arange(len(metrics))
    width = 0.2  # Width of each bar
    
    for i, (model_name, mean_scores) in enumerate(all_mean_scores.items()):
        plt.bar(x + i * width, list(mean_scores.values()), width, label=model_name)
    
    plt.title("Comparison of Models Using Mean Scores")
    plt.xlabel("Metrics")
    plt.ylabel("Mean Score")
    plt.ylim(0, 1)
    plt.xticks(x + width * (len(all_mean_scores) - 1) / 2, metrics)
    plt.legend(title="Models")
    
    # Adding numerical values on top of bars
    for i, (model_name, mean_scores) in enumerate(all_mean_scores.items()):
        for j, value in enumerate(mean_scores.values()):
            plt.text(x[j] + i * width, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=8)
    
    plt.tight_layout()
    plt.show()

file_paths = [
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\Bagging_best_hyperparameters_results.json",
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\RandomForrest_best_hyperparameters_results.json",
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\XGBoost_best_hyperparameters_results.json",
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\SVM_best_hyperparameters_results.json",
]

model_names = ["Random Forest", "Bagging", "XGBoost", "SVM"]

compare_models(file_paths, model_names)
```

```python
# Function to load yearly metrics from a file
def load_yearly_metrics(file_path):
    with open(file_path, "r") as file:
        results = json.load(file)
    years = [result["year"] for result in results]
    metrics = ["accuracy", "precision", "recall", "f1", "weighted_f1"]
    metrics_over_years = {metric: [result[metric] for result in results] for metric in metrics}
    return years, metrics_over_years

# Function to visualize metrics over the years for all models
def plot_metrics_over_years_all_models(file_paths, model_names):
    metrics = ["accuracy", "precision", "recall", "f1", "weighted_f1"]
    all_years = set()  # Collect all years to ensure consistent x-axis
    models_metrics = {}

    # Load data for each model
    for file_path, model_name in zip(file_paths, model_names):
        years, metrics_over_years = load_yearly_metrics(file_path)
        all_years.update(years)
        models_metrics[model_name] = metrics_over_years

    all_years = sorted(all_years)

    # Plot each metric
    for metric in metrics:
        plt.figure(figsize=(10, 6))
        for model_name, metrics_over_years in models_metrics.items():
            years = sorted(models_metrics[model_name][metric])
            plt.plot(
                all_years,
                metrics_over_years[metric],
                marker='o',
                label=model_name
            )
        plt.title(f"{metric.capitalize()} Over Years")
        plt.xlabel("Year")
        plt.ylabel(metric.capitalize())
        plt.grid(True)
        plt.legend(title="Models")
        plt.tight_layout()
        plt.show()

file_paths = [
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\Bagging_best_hyperparameters_results.json",
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\RandomForrest_best_hyperparameters_results.json",
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\XGBoost_best_hyperparameters_results.json",
    "C:\\Users\\hasee\\Desktop\\NDVI-Based Crop Classification for Rice and Cotton Using Machine Learning\\supervisedModels\\Results\\SVM_best_hyperparameters_results.json",
]

model_names = ["Random Forest", "Bagging", "XGBoost", "SVM"]

plot_metrics_over_years_all_models(file_paths, model_names)
```
